# Housing Prices in USA

This data gives different sales prices with respect to type of houses in USA.
We have 5 Predictor variables: Avg. Area Income, Avg. Area House Age, Avg. Area Number of Rooms, Avg. Area Number of Bedrooms, Area Population and
Price as target variable.
We would like to estimate the price of the houses based on different values.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

usa_housing = pd.read_csv("C:\\Users\\Priyanka.Rawat\\Desktop\\Statistics\\Data Science\\DataBase\\USA_Housing.csv")
```

As you can see above, features like age, rooms and bedrooms should be __int__ not __float__, so we will convert these variables to __int__ to make more sense of the data.

```python
usa_housing["Avg. Area Number of Rooms"] = usa_housing["Avg. Area Number of Rooms"].astype("int")
usa_housing["Avg. Area Number of Bedrooms"] = usa_housing["Avg. Area Number of Bedrooms"].astype("int")
usa_housing["Avg. Area House Age"] = usa_housing["Avg. Area House Age"].astype("int")
usa_housing.head(2)
```

```python
usa_housing.drop("Address", axis = 1, inplace = True)
usa_housing.head(2)
```

# Model Development

```python
from sklearn.linear_model import LinearRegression
lm = LinearRegression()
X = usa_housing[["Avg. Area Income","Avg. Area House Age", "Avg. Area Number of Rooms","Avg. Area Number of Bedrooms", "Area Population"]]
y = usa_housing["Price"]
lm.fit(X, y)
```

We will scale the data since the entries in features belong to different range, this creates harmony among them.

```python
from sklearn import preprocessing
pre_process = preprocessing.StandardScaler()
X = pre_process.fit_transform(X)
lm.fit(X,y)
X
```

```python
lm.predict(X)
```

# Train test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=101)
normalized_X = preprocessing.normalize(X)
```
